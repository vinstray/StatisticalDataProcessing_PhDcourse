### 2. Mahalanobis distance
The Mahalanobis distance is a measure of the distance between a point and a distribution. It is a multi-dimensional generalization of the concept of distance that takes into account the correlations between variables.
The Mahalanobis distance is commonly used in statistical classification and anomaly detection, as it provides a way to compare observations in a multivariate space while taking into account the correlations between variables. It is also used in quality control and multivariate analysis to detect outliers and assess the similarity of observations.

MD is quite effective to find outliers for multivariate data. Especially, if there are linear relationships between variables, MD can figure out which observations break down the linearity. 


Let's use airquality data
- Finding the center point of “Ozone” and “Temp”.
- Calculating the covariance matrix of “Ozone” and “Temp”.
- Finding the Mahalanobis Distance of each point to center.
- Finding the Cut-Off value from Chi-Square distribution
- Selecting the distances which is less than Cut-Off (These are the values which isn’t an outlier).

```r
airquality
```

```r
# Select only Ozone and Temp variables
air         = airquality[c("Ozone" , "Temp")]

# We need to remove NA from data set
air         = na.omit(air)

# Finding the center point 
air.center  = colMeans(air)

# Finding the covariance matrix
air.cov     = cov(air)
```

Data plot as ellipse.
Center represents the mean values of variables, shape represents the covariance matrix and radius should be the square root of Chi-Square value with 2 degrees of freedom and 0.95 probability.

```r
# Ellipse radius from Chi-Sqaure distrubiton
rad  = qchisq(p = 0.95 , df = ncol(air))

# Square root of Chi-Square value
rad  = sqrt(rad)

# Finding ellipse coordiantes
ellipse <- car::ellipse(center = air.center , shape = air.cov , radius = rad ,
                        segments = 150 , draw = FALSE)
```

```r
# Call the package
library(ggplot2)

# Ellipse coordinates names should be same with air data set
ellipse <- as.data.frame(ellipse)
colnames(ellipse) <- colnames(air)

# Create scatter Plot
figure <- ggplot(air , aes(x = Ozone , y = Temp)) +
       geom_point(size = 2) +
       geom_polygon(data = ellipse , fill = "orange" , color = "orange" , alpha = 0.5)+
       geom_point(aes(air.center[1] , air.center[2]) , size = 5 , color = "blue") +
       geom_text( aes(label = row.names(air)) , hjust = 1 , vjust = -1.5 ,size = 2.5 ) +
       ylab("Temp Values") + xlab("Ozone Values")
       
# Run and display plot
figure
```

Blue point on the plot shows the center point. Black points are the observations for Ozone — Wind variables. As you can see, the points 30, 62, 117, 99 are outside the orange ellipse. It means that these points might be the outliers.

```r
# Finding distances
distances <- mahalanobis(x = air , center = air.center , cov = air.cov)

# Cutoff value for ditances from Chi-Sqaure Dist. 
# with p = 0.95 df = 2 which in ncol(air)
cutoff <- qchisq(p = 0.95 , df = ncol(air))

## Display observation whose distance greater than cutoff value
air[distances > cutoff ,]
## Returns : 30. 62. 99. 117. observations
```

We have identified the outliers in our multivariate data. Outliers found 30. 62. 99. 117. observations (rows)
