### 1. How to shows the help of functions?
- use ? for help

```r
# help of some function: 
?plot
```

### 2. How to assign value to a variable?

- use <- 

```r
# Assigns number 0 through 10 to x1
x1 <- 0:10
x1

# Descending order
x2 <- 10:0
x2
```

### 3. Other useful  data entering methods

```r
# Ascending values (duplicates 1:10)
(x3 <- seq(10))

# Specify change in values
(x4 <- seq(30, 0, by = -3))

# c = concatenate (or combine or collect)
x5 <- c(5, 4, 1, 6, 7, 2, 2, 3, 2, 8)
x5
```

### 4. How to import data

```r
# TEXT FILES

# Load a spreadsheet that has been saved as comma-delimited
# text file. Need to give complete address to file. This
# command gives an error on missing data but works on
# complete data.
r_txt <- read.table("UWBrangesErrors.txt", header=TRUE)
r_txt
```

```r
#
# READ.CSV FOR CSV FILES ###################################

# R's built-in function for csv files (also used by rio)

# CSV FILES
# Don't have to specify delimiters for missing data
# because CSV means "comma separated values"
r_csv <- read.csv("UWBrangesErrors.csv", header = TRUE)
r_csv
```

```r
# Use pacman to load add-on packages as desired
pacman::p_load(pacman, rio) 
```

```r
#  CSV
rio_csv <- import("UWBrangesErrors.csv")
head(rio_csv)

# TXT
rio_txt <- import("UWBrangesErrors.txt")
head(rio_txt)

# Excel XLSX
rio_xlsx <- import("UWBrangesErrors.xlsx")
head(rio_xlsx)

View(rio_txt)
```
