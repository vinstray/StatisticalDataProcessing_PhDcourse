```r
# Load the UWBranges.csv data file
ranges <- read.csv("observations_def.csv", sep = ";")
# Clean ranges dataset of 0.0 values
ranges[ranges == 0] <- NA
class(ranges)
head(ranges)
class(ranges$X1m)
head(ranges$X1m)
```

```r
#
plot(ranges)
```

```r
#histogram by group (features, speaces, phenomena)

#Put graphs in 5 rows and 2 columns

par(mfrow = c(5,2))

# Histograms for each species using options
hist(ranges$X1m,
  breaks = 9,
  main = "Petal Width for Setosa",
  xlab = "",
  col = "red")

# Histograms for each species using options
hist(ranges$X2m,
  breaks = 9,
  main = "Petal Width for Setosa",
  xlab = "",
  col = "red")

# Histograms for each species using options
hist(ranges$X3m,
  breaks = 9,
  main = "Petal Width for Setosa",
  xlab = "",
  col = "red")

# Histograms for each species using options
hist(ranges$X4m,
  breaks = 9,
  main = "Petal Width for Setosa",
  xlab = "",
  col = "red")
```

```r
#Lets group the data by acquisition distance


# Load the ggplot2 library
library(ggplot2)
library(dplyr)

# Group the data frame by the `gear` column
grouped_data <- ranges %>% group_by(ranges[)

# Calculate the mean value for each group
mean_by_group <- grouped_data %>% summarize(mean = mean(mpg))

#
```
